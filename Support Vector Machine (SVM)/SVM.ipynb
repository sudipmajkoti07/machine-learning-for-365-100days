```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
dir(iris)
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)  # here create the dataframe for our easy ness
```

```python
df['target']=iris.target # here we add the target column  we add target to EDA only
```

```python
df = df.rename(columns={'sepal length (cm)': 'sepal length', 'sepal width (cm)': 'sepal width',"petal length (cm)":"petal length",
                       "petal width (cm)":"petal width"})
```

```python
#we change column name for easyness
```

```python
df
```

```python
df[df['target']==1].head()
```

```python
def add(x):
    if x==1:
        return "versicolor"
    elif x==0:
        return "setosa"
    else:
        return "virginica"

df['flowername']=df.target.apply(add)
# here i learn to use the one column to create the new column using the functions as well.
```

```python
df 
```

```python
df0=df[df['target']==0]
df1=df[df['target']==1]
df2=df[df['target']==2]
# here we give the each flower a different dataframe
```

```python
sns.scatterplot(x=df0['sepal length'], y=df0['sepal width'],color="green",marker="+")
sns.scatterplot(x=df1['sepal length'], y=df1['sepal width'],color="blue")
# here is of sepal
```

```python
sns.scatterplot(x=df0['petal length'], y=df0['petal width'],color="green",marker="+")
sns.scatterplot(x=df1['petal length'], y=df1['petal width'],color="blue")

#Here is of petal
```

```python
from sklearn.model_selection import train_test_split
```

```python
df
```

```python
x=df.drop(['target','flowername'],axis=1)  # here we remove the two columns for ttraining the data
```

```python
x
```

```python
y=df.target
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict([[2,4,5,1]])   # here we predict the flower by giving the sepal l,w and petal l,w .
```

```python
model.score(x_test,y_test)
```
