```python
df=pd.DataFrame(ram.data) #here we create the dataframe using data
```

```python
df
```

```python
df['target']=ram.target
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('target',axis=1)
```

```python
y=df.target  # 
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2) # here we set the training and test dataset
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test) @ the accuracy is very high
```

```python
ypredicted=model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,ypredicted)
```

```python
cm
```

```python
sns.heatmap(cm,annot=True)
```
